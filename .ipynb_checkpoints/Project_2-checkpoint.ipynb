#### E) Use seaborn library to plot histogram for non-null values of ConvertedSalary column.
#### Use different values of bins and see which gives better representation of results
#### Plot histogram for df_sal_NonNa having bin size as 30 and 500


```python
# Setting default seaborn style
sns.set(color_codes=True)

fig ,(ax1,ax2) = plt.subplots(figsize= (20,18), nrows=2, ncols=1)
# adjusting space between two figures
fig.subplots_adjust(hspace=0.25)

# Plotting bin 30
sns.distplot(df_sal_NonNa['ConvertedSalary'],bins=30, kde=False, ax=ax1,color = 'DarkBlue')
ax1.set_title('Histogram with 30 bins.')


# Plotting bin 500
sns.distplot(df_sal_NonNa['ConvertedSalary'],bins=500, kde=False, ax=ax2,color = 'Darkblue')
ax2.set_title('Histogram with 500 bins.')
```

#### Ques -Describe the results plotted in the Histogram

### Ans - Observations

Looking at both the plots we can observe that the average annual pay recieved by the developers is arround 75000 USD. Also there is gradual decrease in the salary between 75K USD to 100K USD, but after that there are hikes in salary pay between 125K USD to 200K USD and then decrease gradually afterwards. The maximum salary earned by the deveopers is arround 1900K - 2000K USD

## Task 3 - Distribution with Multiple Columns

#### A) For this task we will plot stripplot and see the distribution of Salary and countries
#### Plot stripplot keeping ConvertedSalary on the Y-axis and developers' countries on the X-axis. Use df_sal_NonNa


```python
sns.set()
fig, ax = plt.subplots(figsize=(12,60))
# Creating Dictionary to change it in data frame
df_Country_NonNa_Sal = {}
df_Country_NonNa_Sal['ConvertedSalary'] = df_sal_NonNa['ConvertedSalary']

df_Country_NonNa_Sal['Country'] = df_survey[df_survey['ConvertedSalary'].notnull()]['Country']
# Converting dict in dataframe
df_Country_NonNa_Sal = pd.DataFrame(df_Country_NonNa_Sal)

# Plotting strip plot
sns.stripplot(data=df_Country_NonNa_Sal, x='ConvertedSalary', y='Country', size=15)

ax.set_yticklabels(ax.get_yticklabels(), fontsize=15)

plt.show()
```

#### Ques -  What did you see? Which country has the highest Salary and which falls in the lowest salary segment?

#### Ans - Observations
We see the maximum percentage of higher salary distributions are in the developed countries such as *UK, US,France* etc. They are having the proper distributions of salary(from the lower range to higher range).

Also this distributions enumerate the fact that the under developing countries like *India, Spain, UAE,China* has the distribution in middle levels and lower levels more. Also they have some concentration of higher salaries but still in comparision to US And UK concentration stands small.

The Countries such as *Nepal, Cuba, Afghanistan, Zimbabwe* etc, has the salary distribution at very lower levels. Since they are facing the poverty challenges, we can expect such salary concentrations in such parts of globe.

### Task 3 - B) The plot in section A is not very clear in order to make predictions. So we need to select some countries. Lets take the top 5 countries from where the users’ belong to see their salary distribution. Again plot the stripplot using convertedsalary for the top 5 countries.


```python
df_top_5_countries = df_Country_NonNa_Sal['Country'].value_counts().index[:5].tolist()

# Fetch data for only those countries that are in the top_five_countries list.
df_top_5_countries =  df_Country_NonNa_Sal[df_Country_NonNa_Sal.apply(lambda x : x['Country'] in df_top_5_countries,
                                                                            axis=1)]

# Cleaing up the index mess created
df_top_5_countries = df_top_5_countries.reset_index()

del df_top_5_countries['index']

fig, ax = plt.subplots(figsize=(10, 8))

sns.set()

sns.stripplot(data=df_top_5_countries, y='ConvertedSalary', x='Country', size=10)

ax.set_title("Strip Plot")
plt.show()
```

### Ques - Which are the countries from where most developers who took the survey belong?

The Strip plot is very fast in compare to the Swarm plot and gives an average of the distributions knowledge of the data.
Comparing both we can say that to get quick overview of huge amount of data we must be using strip plot and to get distinction among data

## Task 4 – Impact of Education on Salary

#### For this we will use df_sal_NonNa dataframe and FormalEducation and ConvertedSalary columns

### A) See how Formal Education is related to Salary by displaying in the form of swarmplot

```python
sns.set()
fig, ax = plt.subplots(figsize=(12,10))
df_formal_education = {}
df_formal_education['ConvertedSalary'] = df_sal_NonNa['ConvertedSalary']
df_formal_education['FormalEducation'] = df_survey[df_survey['ConvertedSalary'].notnull()]['FormalEducation']
df_formal_education = pd.DataFrame(df_formal_education)

# sns.stripplot(data=df_formal_education, x='ConvertedSalary', y='FormalEducation', size=15)

sns.swarmplot(data=df_formal_education[:2000], x='ConvertedSalary',  y='FormalEducation',size=10)

ax.set_yticklabels(ax.get_yticklabels(), fontsize=15)

ax.set_title("Swarm Plot Formal Education vs Salary")

plt.show()
```


### Ques- Describe the result of swarmplot for FormalEducation and ConvertedSalary columns. Formal Education of the developers who fetches salary in highest range, which is in in lowest range, Salary distribution of which education is distributed evenly.

### Ans - Observations 
As from the above swarm plot we can see that Formal Education plays a very important role in earning a good salary.
The graph states that the developers who completed the *Bachelor's Degree* are earning very good salary and the variation is evenly distributed. We can conclude this too that earning a *Master's Degree* also fetch users good salary but the lower number of distribution at higher salary, state that there is not necessarily a general trend of earning a great salary after completing Master's.
Primary School students and Professional Degree (JD,MD) developer's are earning lower salary. 

#### Conclusions

#### 1 Highest Range - Bachelor's Degree 

#### 2 Lowest Range - Professional Degree(JD, MD, etc.) / Primary school / No formal education

#### 3 Evenly Distributed - Master's Degree / Secondary School / Other Doctoral degree

## Task 6 – Pie Chart

#### A) Create a Pie chart to display the distribution of developers based on their Formal Education

#### Display their percentages on the Pie-Chart
#### The respective Formal Education should be displayed outside the circle
#### The title of the graph should be ‘Classification of Developers by their Formal Education’

```python
plt.figure(figsize=(12,12))
df_formal_education['FormalEducation'].value_counts().plot.pie(autopct = "%1.1f", fontsize=15,
                                              wedgeprops={"linewidth":2})
plt.ylabel(" ")
plt.title("Classification of Developers by their Formal Education",fontsize =20)
plt.show()
```

## Task - 6
#### B) Create the same Piechart as in Section A) with the following changes

#### Explode the three segments with lowest values

```python
plt.figure(figsize=(9,9))
colors = ['skyblue','blue','purple', 'yellowgreen', 'lightcoral', 'yellow', 'red','grey','brown']
explode_val = [0, 0, 0, 0, 0, 0, 0.5,0.8,1]
df_formal_education['FormalEducation'].value_counts().plot.pie(explode=explode_val, autopct = "%1.1f", colors=colors, fontsize=15,
                                              wedgeprops={"linewidth":3})
plt.ylabel(" ")
plt.title("Classification of Developers by their Formal Education",fontsize=20)
plt.show()
```
